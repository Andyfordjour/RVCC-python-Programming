```python
# for the GUI - callback functions / event handlers
import tkinter as tk

# for adding frames and components to root window
from tkinter import ttk

class MyGui:
    def __init__(self):
        
        # create an empty root window
        self.root = tk.Tk()
        self.root.title("Future Value Calculator")
        self.root.geometry("300x200")

        # adding a frame
        self.frame = ttk.Frame(self.root, padding="10 10 10 10")
        self.frame.pack(fill=tk.BOTH, expand=True)

        # creates two buttons and adds them to the frame
        # command connects the buttons to functions when they are clicked
        button1 = ttk.Button(self.frame, text="Click me!", command=self.click_button1)
        button2 = ttk.Button(self.frame, text="Bye!", command=self.click_button2)

        # display the buttons
        button1.pack()
        button2.pack()

        # make the root window visible
        self.root.mainloop()

        # callback function for button1
        # this is called callback function or event handler
    def click_button1(self):
        self.root.title("You clicked the top button!")
    
    # callback function for button2
    def click_button2(self):
        # closes the root window , causes loop to end which ends program
        self.root.destroy()

my_gui = MyGui()

```

```python
# using grid() for layouts
import tkinter as tk

# for adding frames and components to root window
from tkinter import ttk

# this class inherits a frame
class EmployeeInfo(ttk.Frame):
    def __init__(self, parent):
        
        ttk.Frame.__init__(self, parent, padding="10 10 10 10")
        self.pack(fill=tk.BOTH, expand=True)
        
        # Define string variables for the entry fields
        # for text entry field
        self.name = tk.StringVar()
        self.employerName = tk.StringVar()

        # create a label using grid to layout the components of a frame
        ttk.Label(self, text="Name:").grid(column=0, row=0, sticky=tk.E)
        ttk.Entry(self, width=25, textvariable=self.name).grid(column=1, row=0)
        
        ttk.Label(self, text="Employer Name:").grid(column=0, row = 1, sticky=tk.E)
        ttk.Entry(self, width=25, textvariable=self.employerName).grid(column=1, row=1)
        
        
        # creates two buttons and adds them to the frame
        ttk.Button(self, text="Clear", command=self.clear).grid(column=2, row=0)

        # Add padding to all child components
        for child in self.winfo_children():
            child.grid_configure(padx=5, pady=3)
            
    def clear(self):
        print("Employee Information:", self.name.get())
        print("Employer Information:", self.employerName.get())
        self.name.set("")
        self.employerName.set("")
    

def main():
    root = tk.Tk()
    root.title("Employee Information")
    
    # Create the frame
    EmployeeInfo(root)
    
    # Display the frame
    root.mainloop()
    
    
if __name__ == "__main__":
    main()

```
