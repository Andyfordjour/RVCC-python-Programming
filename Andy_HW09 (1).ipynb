```python

def display_menu():
    """
    Displays the manu
    """
    print('1. Print Phone Numbers')
    print('2. Add a Phone Number')
    print('3. Remove a Phone Number')
    print('4. Lookup a Phone Number')
    print('5. Quit')
    print()

    
def main():
    """
    The main function ask user to select from menu and displays the command 
    """
    phonebook = {"Irene":"212-555-1111",
            "Maria":"973-555-4444",
            "Linda": "732-555-2222",
            "Paula": "908-555-7777"   }
    menu_choice = 0
    display_menu()
    while menu_choice != 5:
        menu_choice = int(input("Type in a number (1-5): "))
        if menu_choice == 1:
            print("Telephone Numbers:")
            for x in phonebook.keys():
                print("Name: ", x, "\tNumber:", phonebook[x])
            print()
            display_menu()
        elif menu_choice == 2:
            print("Add Name and Number")
            name = input("Name: ")
            phone = input("Number: ")
            phonebook[name] = phone
            print(name + " added to phonebook" )
            print()
            display_menu()
        elif menu_choice == 3:
            print("Remove Name and Number")
            name = input("Name: ")
            if name in phonebook:
                del phonebook[name]
            else:
                print(name, "was not found")
            print()
            display_menu()
        elif menu_choice == 4:
            print("Lookup Number")
            name = input("Name: ")
            if name in phonebook:
                print("The number is", phonebook[name])
            else:
                print(name, "was not found")
            print()
            display_menu()
        elif menu_choice != 5:
            print_menu()
            print()
            display_menu()
            
if __name__ == "__main__":
    main()
```
