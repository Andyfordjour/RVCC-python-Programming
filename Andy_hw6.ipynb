```python
# Movie - Phase 2 - Movie Name and Movie Year
# Author: Andy fordjour

def display_menu():
    """
    Displays_menu function displays the menu of the various functions to be executed
    """
    print("COMMAND MENU")
    print("list - List all movies")
    print("year - year of release")
    print("add - Add a movie")
    print("del - Delete a movie")
    print("exit - exit program")
    
def list(movie_list):
    """
    The list function displays the list of contacts by their first names
    """
    if len(movie_list) == 0:
        print("There are no movies in the list.\n")
    else:
        i = 1
        print()
        for row in movie_list:
            print(str(i) + ". " + row[0] + " (" + str(row[1]) + ")")
            i += 1
        
        print()
        
def year(movie_list):
    """
    The year function ask user for the movie year and and displays the various movies in that year category
    """
    
    year_released = int(input("year: "))
    print()
    if len(movie_list) == 0:
            print("no movie found")                    
    else:            
        for movie in movie_list:
            for year in movie:
                if year_released == year:
                    print(movie[0] + " (" + str(movie[1]) + ")")
    print()
    
def add(movie_list):
    """
    The add function add a new contact to the contact list
    """
    name = input("Movie Name: ")
    year = input("Movie Year: ")
    price = input("Movie Price: ")
    movie = []
    movie.append(name)
    movie.append(year)
    movie.append(price)
    movie_list.append(movie)
    print(movie[0] + " was added.\n")
    
def delete(movie_list):
    """
    The delete function ask for the contact number and delect the contact from the list
    """
    number = int(input("Movie number: "))
    if number < 1 or number > len(movie_list):
        print("Invalid movie number.\n")
    else:
        movie = movie_list.pop(number-1)
        print(movie[0] + " was deleted.\n")
              
def main():
    """
    The main function ask user to select from menu and displays the command 
    """
    movie_list = [["A Star is Born", 2018,4.99],
                     ["An Old Man and a Gun", 2018,5.99],
                    ["The Spy Who Dumped Me", 2017,2.99]]
              
    display_menu()
    command = input("Enter Command: ")
        
    while (command.lower() != "exit"):
        
        if command.lower() == "list":
            list(movie_list)
            display_menu()
        elif command.lower() == "add":
            add(movie_list)
            display_menu()
        elif command.lower() == "del":
            delete(movie_list)
            display_menu()
        elif command.lower() == "year":
            year(movie_list)
            display_menu()
        else:
            print("Not a valid command.  Please try again.\n")

        command = input("Enter Command: ")
         
    print("You'll be back!")     
        
if __name__ == "__main__":
    main()
  
```
