```python
#Author Andy fordjour
# the movie class and object lab
class Movie:
    # constructor that initializes three attributes
    def __init__(self, name,year):
        self.__name = name
        self.__year = year
    #setters for the private data members
    def setName(self,name):
        self.__name = name
        
    def setYear(self,year):
        self.__year = year
    #getters for the private data members 
    def getName(self):
        return self.__name
    
    def getYear(self):
        return self.__year
        
    #getStr returns the name of the movie followd by the year
    def getStr(self):
        
        return self.__name + "("+ str(self.__year)+ ")"
                
def display_menu():
    """Displays the manu"""
    print("COMMAND MENU")
    print("list - List all movies")
    print("add - Add a movie")
    print("del - Delete a movie")
    print("exit - exit program")
    print()
        
def add(movie_list):
    """This functionadds a new movie to the list with inputs from the user
    """
    name = input("Movie Name: ")
    year = input("Movie Year: ")
    movie = Movie(name,year)
    movie_list.append(movie)
    print(movie.getStr() + " was added.\n")

    print()
    
def list(movie_list):
    """The funtion displays the list of movies and the year 
    """
    if len(movie_list) == 0:
        print("There are no movies in the list.\n")
    else:
        i = 1
        print()
        for movie in movie_list:
            print(str(i) + "."+movie.getStr())
            i += 1   
    print()

def delete(movie_list):
    """
    The delete function ask for the contact number and delect the contact from the list
    """
    number = int(input("Movie number: "))
    if number < 1 or number > len(movie_list):
        print("Invalid movie number.\n")
    else:
        movie = movie_list.pop(number-1)
        print(movie.getStr() + " was deleted.\n")
    print()
    
def main():
    """
    The main function ask user to select from menu and displays the command 
    """
    movie_list = [Movie("A Star is Born", 2018),
                Movie("An Old Man and a Gun", 2018),
                Movie("The Spy Who Dumped Me", 2017)]
              
    display_menu()
    command = input("Enter Command: ")
        
    while (command.lower() != "exit"):
        
        if command.lower() == "list":
            list(movie_list)
            #print()
            display_menu()
        elif command.lower() == "add":
            add(movie_list)
           # print()
            display_menu()
        elif command.lower() == "del":
            delete(movie_list)
           # print()
            display_menu()
        else:
            print("Not a valid command.  Please try again.\n")

        command = input("Enter Command: ")
         
    print("You'll be back!")     
        
if __name__ == "__main__":
    main()
    
        
```
