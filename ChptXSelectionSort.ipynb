```python
#
#  The selectionSort function sorts a list using the selection sort algorithm.
#  Author:     Horstmann and Necaise
#  Updated by: L Yang 11/8/2019
#

import random

## Sorts a list, using selection sort.
#  @param values the list to sort
#
def selectionSort(values) :
    for i in range(len(values)) :
        minPos = minimumPosition(values, i)
        
        # swap the two elements
        temp = values[minPos] 
        values[minPos] = values[i]
        values[i] = temp

## Finds the smallest element in a tail range of the list.
#  @param values the list to sort
#  @param start the first position in values to compare
#  @return the position of the smallest element in the
#  range values[start] . . . values[len(values) - 1]
#
def minimumPosition(values, start) :
    minPos = start
    for i in range(start + 1, len(values)) :
        if values[i] < values[minPos] :
            minPos = i
         
    return minPos


##
#  The main function  the selection sort algorithm by sorting a
#  list that is filled with random numbers.
def main():
    ''' The main fuctiom to invoke selection sort with  list
        filled with random numbers
    '''
    n = 20
    values = []
    for i in range(n) :
        values.append(random.randint(1, 100))   
    print(values)
    selectionSort(values)
    print(values)

if __name__ == "__main__":
    main()
```

```python
# Sort with timer
# Author:    Horstmann and Necaise
# Upated by: L Yang 11/8/2019

import time as t
import random

## Sorts a list, using selection sort algorithm
#  @param values the list to sort
#
def selectionSort(values) :
    for i in range(len(values)) :
        minPos = minimumPosition(values, i)
        # swap the two elements
        temp = values[minPos]
        values[minPos] = values[i]
        values[i] = temp

## Finds the smallest element in a tail range of the list.
#  @param values the list to sort
#  @param start the first position in values to compare
#  @return the position of the smallest element in the
#  range values[start] . . . values[len(values) - 1]
#
def minimumPosition(values, start) :
    minPos = start
    for i in range(start + 1, len(values)) :
        if values[i] < values[minPos] :
            minPos = i
         
    return minPos

##
#  The main function measures how long it takes to sort a list of a 
#  user-specified size with the selection sort algorithm.
#
def main():
    firstSize = int(input("Enter first list size: "))
    numberOfLists = int(input("Enter number of lists: "))

    for k in range(1, numberOfLists + 1) :
        size = firstSize * k
        values = []
        # Construct random list.
        for i in range(size) :
            values.append(random.randint(1, 100))   

        startTime = t.time()
        selectionSort(values)
        endTime = t.time()
      
        print("Size: %d Elapsed time: %.3f seconds" % (size, endTime - startTime))
    
if __name__ == "__main__":
    main()
```

```python
# Linear Search
# Author:     Horstmann and Necaise
# Updated by: L Yang 11/8/2019

##
#  This module implements a function for executing linear searches in a list.
#

import random

## Finds a value in a list, using the linear search algorithm.
#  @param values the list to search
#  @param target the value to find
#  @return the index at which the target occurs, or -1 if it does not 
#  occur in the list
#
def linearSearch(values, target) :
    for i in range(len(values)) :
        if values[i] == target :
            return i
         
    return -1

##
#  The main function demonstrates the linear search algorithm.
#
def main():
    
    # Construct random list.
    done = False
    n = 20
    values = []
    for i in range(n) :
        values.append(random.randint(1, 20))   
    print(values)

    while not done :
        target = int(input("Enter number to search for, -1 to quit: "))
        if target == -1 :
            done = True
        else :
            pos = linearSearch(values, target)
            if pos == -1 :
                print("Not found")
            else :
                print("Found in position", pos) 

if __name__ == "__main__":
    main()
```
