Problem 2: Dog Class

```python
# Student in Class
# This is missing docstrings in some of the methods!!!
# Dog class

class Dog:

    def __init__(self, name = "", breed = ""):
            self.__name = name
            self.__breed = breed
            self.__tricks = []
            
    def setName(self, name):
        self.__name = name
        
    def getName(self):
        return self.__name
    
    def setBreed(self, breed):
        self.__breed = breed
        
    def getBreed(self):
        return self.__breed
    
    def teach(self, trick):
        """Teaches the dog object a trick"""
        
        self.__tricks.append(trick)
        
        print(self.__name + " knows " + trick)
        print()
            
        
    def knows(self, trick):
        """Determines if dog object knows a trick"""
        
        if trick in self.__tricks:
            print("Yes, " + self.__name + " knows " + trick)
            print()
        else:
            print("No, " + self.__name + " does not know " + trick)
            print()
            
def main():
    """This is the main"""
    
    dog1 = Dog("Teddie", "German Shepherd")
    
    print("Your dogs name is " + dog1.getName() + " and he is a " + dog1.getBreed())
    print()
    
    dog1.teach("frisbee")
    dog1.teach("fetch")
    dog1.teach("rollover")
    
    dog1.knows("rollerblading")
    dog1.knows("fetch")
    
if __name__ == "__main__":
    main()    
        
```
