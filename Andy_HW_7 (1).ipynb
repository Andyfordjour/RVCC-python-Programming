```python
# this program is to take a csv file of a monthly sales and displays the total and average using the various functions
import csv

def readFile():
    """Method that reads data from CSV file and returns list of sales"""

    sales = [] 
    with open("monthly_sales.csv") as file:
        reader = csv.reader(file)
        for row in reader:
            sales.append([row[0],row[1]])
        return sales

def writeToFile(sales):
    """Method that takes  list of sales and writes it to file"""

    with open("monthly_sales.csv","w+") as outFile:
        for sale in sales:
            outFile.write(f"{sale[0]},{sale[1]}\n")

def monthly(sales):
    """monthly displays the list in the file """
    with open("monthly_sales.csv", newline="") as file:
        reader = csv.reader(file)
        for sale in sales:
            print(sale[0]+" - "+ sale[1])
        print()
        
def yearly(sales):
    """Function to displaye total of yearly sales
    and monthly average sales"""
    total = 0

    for sale in sales:
        try:
            total += int(sale[1])
        except ValueError:
            print("Using sales amount of 0 for month",sale[0])
        total += 0

    average = total / len(sales)
    print("Yearly total: ",total)
    print("Monthly average: ",round(average,2))
    print()
    
def edit(sales):
    """Function to edit sales of a given month ad re-write to file"""

    monthsList = [sale[0].lower() for sale in sales]

    month = input("Three-letter month : ")

    if month.lower() not in monthsList:

        print("Invalid three-letter month.")
    else:

        salesAmount = int(input("Sales Amount: "))

    for sale in sales:
        if month == sale[0].lower():
            sale[1] = str(salesAmount)

    writeToFile(sales)
    print("Sales amount for",month,"was modified")
    print()

def display_menu():
    """display function displays the menu
    """
    print("Monthly Sales program")
    print("COMMAND MENU")
    print("monthly - View monthly sales")
    print("yearly - View yearly summary")
    print("edit - Edit sales for a month")
    print("exit - Exit program")
    print()

def main():
    """The main function ask for the command and execute the various
    """
    sales = readFile()

    writeToFile(sales)
    display_menu()

    command = input("Enter Command: ")
        
    while (command.lower() != "exit"):
        if command.lower() == "monthly":
            monthly(sales)
            display_menu()
            command = input("Enter Command: ")
        elif command.lower() == "yearly":
            yearly(sales)
            display_menu()
            command = input("Enter Command: ")
        elif command.lower() == "edit":
            edit(sales)
            display_menu()
            command = input("Enter Command: ")
        else:
            print("Not a valid command.  Please try again.\n")
            command = input("Enter Command: ")

    print("Bye!")     
#calls main method
if __name__ == "__main__":
    main()
```

```python
#program to draw a simple  population bar chart

import matplotlib.pyplot as plt

def main():
    """
    code to illustrate population growth in barplot at some years interval (1995-2015)
    """
    # Create a list with the X coordinates of each bar's left edge
    left_edges = [1995, 2000, 2005, 2010,2015]
    
    # Create a list iwht the heights of each bar
    heights = [266.28,282.16,295.52,309.32,320.64]
    
    # Create a variable for the bar width
    bar_width = 2
    
    # Add a title
    plt.title('Population by Year ')
    
    # Add labels to the axes
    plt.xlabel('Year')
    plt.ylabel('population in million')
    
    # Customize the tick marks
    
    # Build the bar chart
    plt.bar(left_edges, heights, bar_width, 
            color=['magenta', 'black', 'blue', 'red', 'yellow'])
    
    # Display the bar chart
    plt.show()
    
if __name__ == "__main__":
    main()
```
