```python

def display_menu():
    """
    Displays_menu function displays the menu of the various functions to be executed
    """
    print("COMMAND MENU")
    print("list - Display all contacts")
    print("view - View a contact")
    print("add - Add a contact")
    print("del - Delete a contact")
    print("exit - Exit program")
    
def list(contact_list):
    """
    The list function displays the list of contacts by their first names
    """
    if len(contact_list) == 0:
        print("There are no contacts in the list.\n")
    else:
        i = 1
        print()
        for row in contact_list:
            print(str(i) + ". " + row[0] )
            i += 1
        
        print()
        
def view(contact_list):
    """
    The view list ask for the contact number and displays all informtion about the contact 
    """
    number = int(input("number: "))
    if number < 1 or number > len(contact_list):
        print("Invalid number.\n")
    else:
        results = contact_list.pop(number-1)
        print("Name: ",results[0])
        print("Email: ",results[1])
        print("Phone:",results[2])    
    print()
              
    
def add(contact_list):
    """
    The add function add a new contact to the contact list
    """
    name = input("Name: ")
    email = input("email: ")
    phone = input("Phone: ")
    contact = []
    contact.append(name)
    contact.append(email)
    contact.append(phone)
    contact_list.append(contact)
    print(contact[0] + " was added.\n")
    
def delete(contact_list):
    """
    The delete function ask for the contact number and delect the contact from the list
    """
    number = int(input("number: "))
    if number < 1 or number > len(contact_list):
        print("Invalid contact number.\n")
    else:
        contact = contact_list.pop(number-1)
        print(contact[0], " was deleted.\n")
              
def main():
    """
    The main function ask user to select from menu and displays the command 
    """
    print("Welcome to Contact Manager!\n")
              
    display_menu()
    print()
    command = input("Enter Command: ")
    
    contact_list = [["Linda Yang","linda@yahoo.com","908-322344"],
                    ["Michele VanBuren","michele@verizon.net","+442079460958"]]     
    while (command.lower() != "exit"):
        
        if command.lower() == "list":
            list(contact_list)
            
            #display_menu()
        elif command.lower() == "add":
            add(contact_list)
            
            #display_menu()
        elif command.lower() == "del":
            delete(contact_list)
            #display_menu()
            
        elif command.lower() =="view":
            view(contact_list)
            #display_menu()
            
        else:
            print("Not a valid command.  Please try again.\n")
            
        command = input("Enter Command: ")
               
    print("Bye!")     
        
if __name__ == "__main__":
    main()
  
```
