```python
def debate_file(filename):
    """
    the debate_file function takes the file split and sort accordingly 
    """
    
    with open(filename)as file:
        text = file.read()
        
    text = text.replace("\n"," ")
    text = text.replace(",", "")
    text = text.replace(".","")
    
    words =text.split(" ")
    words.sort()
    return(words)

def get_most_words(words):
    """
    the get most words functions counts the words in the text file
    """
    most_words ={}
   # most_words.append(words[0])
    for word in words:
        if word in most_words:
            most_words[word] += 1
        else:
            most_words[word] = 1
    return most_words
    
def display_word(most_words):
    """
    The display function displays the the most words and the most word occurances
    """
    words =list(most_words.keys())
    words.sort(key=str.lower)
    for word in words:
        count = most_words[word]
        print(word,"=",count)
        
    
def main():
    """
    the main function execute the various functions
    """
    filename = "DebateTranscript.txt"
    words = debate_file(filename)
    most_words = get_most_words(words)
    
    print("Number of words =", len(words))
    print("Number of most_words =",len(most_words))
    print("most word occurances:")
    display_word(most_words)
    
    
    
        
if __name__ =="__main__":
    main()
    
    
```
